```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

##nltk.download('punkt')
##nltk.download('stopwords')
##nltk.download('wordnet')
nltk.download('omw-1.4')

text = "Is there a way to fix the muleteer check during the moratorium"
```

```python
## CREATE TOKENS W SENTENCES
sentences = sent_tokenize(text)
```

```python
## CREATE TOKENS W WORKS
words = [word_tokenize(sent) for sent in sentences]
```

```python
## REMOVING STOPWORDS

## lib with stopwords
from nltk.corpus import stopwords

## punctuation symbols
from string import punctuation

## my own list of stop words
custom_stopwords = list(["seba"])

all_stopwords = set(stopwords.words('english')+list(punctuation)+custom_stopwords)

words_wo_stopwords = [word for word in word_tokenize(text) if word not in all_stopwords]

print(words_wo_stopwords)
```

```python
from nltk.collocations import *

bigram_measures = nltk.collocations.BigramAssocMeasures()

finder = BigramCollocationFinder.from_words(words_wo_stopwords)

print(finder.ngram_fd.items())
```

```python
## STEMMING

from nltk.stem.lancaster import LancasterStemmer
st = LancasterStemmer()

stemmed_wo_stopwords = [st.stem(word) for word in words_wo_stopwords]

print(stemmed_wo_stopwords)
```

```python
## DISAMBIGUATING WORD MEANING

from nltk.corpus import wordnet as wn

for ss in wn.synsets('bass'):
  print (ss, ss.definition())
```

```python
from nltk.wsd import lesk

sense1 = lesk(word_tokenize('this sea bass was really hard to catch'), 'bass')

print(sense1, sense1.definition())
```
